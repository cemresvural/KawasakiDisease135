```python
df1=pd.DataFrame(df)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df1)
```

```python
scores = []
for k in range(2,15):
    km = KMeans(n_clusters=k,random_state=123)
    km = km.fit(df1)
    scores.append(km.inertia_)
dfk = pd.DataFrame({'Cluster':range(2,15), 'Score':scores})
plt.figure(figsize=(8,5))
plt.plot(dfk['Cluster'], dfk['Score'], marker='o')
plt.xlabel('KÃ¼me SayÄ±sÄ±')
plt.ylabel('Inertia')
plt.show()
```

```python
from sklearn import metrics
```

```python
for i in range(4,10):
    kmeans_labels=KMeans(n_clusters=i,random_state=123).fit_predict(df1)
    print("Silhouette skoru {} iÃ§in  clusters k-means : {} ".format(i,metrics.silhouette_score(df1,kmeans_labels, metric='euclidean')))
```

```python
for i in [4,5,6,7,8,9]:
    kmeans_labels=KMeans(n_clusters=i,random_state=123).fit_predict(df1)
    print('Davies Bouldin Score:'+str(metrics.davies_bouldin_score(df1,kmeans_labels).round(3)))
```

```python
kmeans_labels=KMeans(n_clusters=4,random_state=123).fit_predict(df1)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3).fit_transform(df1)
fig = plt.figure(figsize=(20, 9), dpi=80, facecolor='w', edgecolor='k')
ax = plt.axes(projection="3d")
ax.scatter3D(pca.T[0],pca.T[1],pca.T[2],c=kmeans_labels,cmap='Spectral')

xLabel = ax.set_xlabel('X')
yLabel = ax.set_ylabel('Y')
zLabel = ax.set_zlabel('Z')
```
